```python
cur_x=3
rate = 0.01
precision = 0.000001
previous_state_size=1
max_iters = 1000
iters = 0

def df(x):
    return (2*(x+5))
```

```python
while previous_state_size > precision and iters<max_iters:
    prev_x = cur_x
    cur_x = cur_x-rate*df(prev_x)
    previous_state_size = abs(cur_x-prev_x)
    iters=iters+1
    print("Iteration is: ",iters,"\nX value is ",cur_x)
    print("The local minima occurs at ",cur_x)
```
